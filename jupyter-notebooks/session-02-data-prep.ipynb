### 1.2 Comprehensive Data Quality Assessment

Let's create a comprehensive data quality assessment function:

```python
def comprehensive_data_quality_report(df, dataset_name="Dataset"):
    """
    Generate a comprehensive data quality report
    """
    print(f"{'='*60}")
    print(f"DATA QUALITY REPORT: {dataset_name.upper()}")
    print(f"{'='*60}")
    
    # Basic Information
    print(f"\nüìä BASIC INFORMATION")
    print(f"   Rows: {df.shape[0]:,}")
    print(f"   Columns: {df.shape[1]}")
    print(f"   Memory Usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
    
    # Data Types
    print(f"\nüè∑Ô∏è  DATA TYPES")
    dtype_counts = df.dtypes.value_counts()
    for dtype, count in dtype_counts.items():
        print(f"   {dtype}: {count} columns")
    
    # Missing Values Analysis
    print(f"\n‚ùå MISSING VALUES ANALYSIS")
    missing_stats = df.isnull().sum()
    missing_percent = (missing_stats / len(df)) * 100
    
    missing_df = pd.DataFrame({
        'Missing_Count': missing_stats,
        'Missing_Percentage': missing_percent
    }).sort_values('Missing_Percentage', ascending=False)
    
    columns_with_missing = missing_df[missing_df['Missing_Count'] > 0]
    if len(columns_with_missing) > 0:
        print("   Columns with missing values:")
        for col, row in columns_with_missing.iterrows():
            print(f"   {col}: {row['Missing_Count']} ({row['Missing_Percentage']:.1f}%)")
    else:
        print("   ‚úÖ No missing values found!")
    
    # Duplicate Analysis
    print(f"\nüîÑ DUPLICATE ANALYSIS")
    total_duplicates = df.duplicated().sum()
    duplicate_percentage = (total_duplicates / len(df)) * 100
    print(f"   Total duplicates: {total_duplicates} ({duplicate_percentage:.2f}%)")
    
    # Unique Values Analysis
    print(f"\nüéØ UNIQUENESS ANALYSIS")
    for col in df.columns:
        unique_count = df[col].nunique()
        unique_percentage = (unique_count / len(df)) * 100
        if unique_percentage < 1:
            print(f"   {col}: {unique_count} unique values ({unique_percentage:.2f}%) - Potentially categorical")
        elif unique_percentage > 95:
            print(f"   {col}: {unique_count} unique values ({unique_percentage:.2f}%) - Likely identifier")
    
    # Outlier Analysis (for numerical columns)
    print(f"\nüìà OUTLIER ANALYSIS (IQR Method)")
    numerical_cols = df.select_dtypes(include=[np.number]).columns
    
    for col in numerical_cols:
        if df[col].dtype in ['int64', 'float64'] and df[col].nunique() > 10:
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
            outlier_percentage = (len(outliers) / len(df)) * 100
            
            if len(outliers) > 0:
                print(f"   {col}: {len(outliers)} outliers ({outlier_percentage:.1f}%)")
                print(f"      Range: [{df[col].min():.2f}, {df[col].max():.2f}]")
                print(f"      Expected range: [{lower_bound:.2f}, {upper_bound:.2f}]")
    
    return missing_df, dtype_counts
```

### 1.3 Visual Data Quality Dashboard

```python
def create_data_quality_dashboard(df, dataset_name="Dataset"):
    """
    Create a visual dashboard for data quality assessment
    """
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle(f'Data Quality Dashboard: {dataset_name}', fontsize=16, fontweight='bold')
    
    # 1. Missing Values Heatmap
    missing_data = df.isnull()
    sns.heatmap(missing_data, yticklabels=False, cbar=True, cmap='viridis', ax=axes[0,0])
    axes[0,0].set_title('Missing Values Pattern')
    axes[0,0].set_xlabel('Columns')
    
    # 2. Missing Values Bar Chart
    missing_counts = df.isnull().sum()
    missing_counts = missing_counts[missing_counts > 0].sort_values(ascending=True)
    if len(missing_counts) > 0:
        missing_counts.plot(kind='barh', ax=axes[0,1], color='coral')
        axes[0,1].set_title('Missing Values Count by Column')
        axes[0,1].set_xlabel('Number of Missing Values')
    else:
        axes[0,1].text(0.5, 0.5, 'No Missing Values', ha='center', va='center', fontsize=14)
        axes[0,1].set_title('Missing Values Count by Column')
    
    # 3. Data Types Distribution
    dtype_counts = df.dtypes.value_counts()
    axes[0,2].pie(dtype_counts.values, labels=dtype_counts.index, autopct='%1.1f%%', startangle=90)
    axes[0,2].set_title('Data Types Distribution')
    
    # 4. Numerical Columns Distribution
    numerical_cols = df.select_dtypes(include=[np.number]).columns
    if len(numerical_cols) > 0:
        df[numerical_cols].hist(bins=20, ax=axes[1,0], alpha=0.7)
        axes[1,0].set_title('Numerical Columns Distribution')
    else:
        axes[1,0].text(0.5, 0.5, 'No Numerical Columns', ha='center', va='center', fontsize=14)
        axes[1,0].set_title('Numerical Columns Distribution')
    
    # 5. Categorical Columns Analysis
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns
    if len(categorical_cols) > 0:
        cat_uniqueness = [df[col].nunique() for col in categorical_cols[:5]]  # Top 5 categorical columns
        axes[1,1].bar(categorical_cols[:5], cat_uniqueness, color='lightblue')
        axes[1,1].set_title('Unique Values in Categorical Columns')
        axes[1,1].set_xlabel('Columns')
        axes[1,1].set_ylabel('Number of Unique Values')
        axes[1,1].tick_params(axis='x', rotation=45)
    else:
        axes[1,1].text(0.5, 0.5, 'No Categorical Columns', ha='center', va='center', fontsize=14)
        axes[1,1].set_title('Categorical Columns Analysis')
    
    # 6. Completeness Score
    completeness_scores = (1 - df.isnull().sum() / len(df)) * 100
    completeness_scores.plot(kind='bar', ax=axes[1,2], color='lightgreen')
    axes[1,2].set_title('Data Completeness by Column (%)')
    axes[1,2].set_xlabel('Columns')
    axes[1,2].set_ylabel('Completeness %')
    axes[1,2].tick_params(axis='x', rotation=45)
    axes[1,2].axhline(y=90, color='red', linestyle='--', alpha=0.7, label='90% Threshold')
    axes[1,2].legend()
    
    plt.tight_layout()
    plt.show()
```

### 2.1 Understanding Missing Data Mechanisms

Before choosing an imputation strategy, we need to understand WHY data is missing:

1. **Missing Completely at Random (MCAR):** Missing values are independent of all variables
2. **Missing at Random (MAR):** Missing values depend on observed variables
3. **Missing Not at Random (MNAR):** Missing values depend on unobserved variables

```python
def analyze_missing_patterns(df):
    """
    Analyze patterns in missing data to understand the mechanism
    """
    import missingno as msno
    
    # If missingno is not installed: pip install missingno
    # For this example, we'll create a simplified version
    
    print("MISSING DATA PATTERN ANALYSIS")
    print("="*50)
    
    # Calculate missing data correlations
    missing_df = df.isnull()
    missing_corr = missing_df.corr()
    
    print("Correlation between missing values in different columns:")
    print("(High correlations suggest patterns in missingness)")
    
    # Find high correlations
    for i in range(len(missing_corr.columns)):
        for j in range(i+1, len(missing_corr.columns)):
            corr_val = missing_corr.iloc[i, j]
            if abs(corr_val) > 0.3:  # Threshold for significant correlation
                col1, col2 = missing_corr.columns[i], missing_corr.columns[j]
                print(f"   {col1} <-> {col2}: {corr_val:.3f}")
    
    # Analyze missing combinations
    missing_combinations = df.isnull().value_counts().head(10)
    print(f"\nTop 10 missing value combinations:")
    for pattern, count in missing_combinations.items():
        pattern_str = ", ".join([f"{col}:{val}" for col, val in zip(df.columns, pattern)])
        print(f"   {count:4d} rows: {pattern_str}")
```

### 2.2 Comprehensive Missing Data Strategies

```python
class MissingDataHandler:
    """
    Comprehensive class for handling missing data with multiple strategies
    """
    
    def __init__(self, df):
        self.df = df.copy()
        self.original_shape = df.shape
        self.strategies_applied = []
    
    def deletion_strategies(self, strategy='listwise'):
        """
        Apply deletion strategies for missing data
        """
        if strategy == 'listwise':
            # Remove rows with any missing values
            cleaned_df = self.df.dropna()
            removed_rows = len(self.df) - len(cleaned_df)
            print(f"Listwise deletion: Removed {removed_rows} rows ({removed_rows/len(self.df)*100:.1f}%)")
            
        elif strategy == 'pairwise':
            # Keep rows for analysis on a case-by-case basis
            print("Pairwise deletion: Analyze each variable pair separately")
            cleaned_df = self.df.copy()
            
        elif strategy == 'column_threshold':
            # Remove columns with > threshold% missing values
            threshold = 0.5  # 50% threshold
            missing_percent = self.df.isnull().sum() / len(self.df)
            cols_to_drop = missing_percent[missing_percent > threshold].index
            cleaned_df = self.df.drop(columns=cols_to_drop)
            print(f"Column threshold deletion: Removed columns {list(cols_to_drop)}")
            
        self.strategies_applied.append(f"Deletion: {strategy}")
        return cleaned_df
    
    def simple_imputation(self, strategy='mean', columns=None):
        """
        Apply simple imputation strategies
        """
        df_imputed = self.df.copy()
        
        if columns is None:
            numerical_cols = df_imputed.select_dtypes(include=[np.number]).columns
            categorical_cols = df_imputed.select_dtypes(include=['object']).columns
        else:
            numerical_cols = [col for col in columns if df_imputed[col].dtype in [np.number]]
            categorical_cols = [col for col in columns if df_imputed[col].dtype == 'object']
        
        # Numerical imputation
        if strategy == 'mean':
            for col in numerical_cols:
                if df_imputed[col].isnull().any():
                    mean_val = df_imputed[col].mean()
                    df_imputed[col].fillna(mean_val, inplace=True)
                    print(f"   {col}: Filled {self.df[col].isnull().sum()} missing values with mean ({mean_val:.2f})")
        
        elif strategy == 'median':
            for col in numerical_cols:
                if df_imputed[col].isnull().any():
                    median_val = df_imputed[col].median()
                    df_imputed[col].fillna(median_val, inplace=True)
                    print(f"   {col}: Filled {self.df[col].isnull().sum()} missing values with median ({median_val:.2f})")
        
        elif strategy == 'mode':
            # For both numerical and categorical
            all_cols = list(numerical_cols) + list(categorical_cols)
            for col in all_cols:
                if df_imputed[col].isnull().any():
                    mode_val = df_imputed[col].mode().iloc[0] if not df_imputed[col].mode().empty else 'Unknown'
                    df_imputed[col].fillna(mode_val, inplace=True)
                    print(f"   {col}: Filled {self.df[col].isnull().sum()} missing values with mode ({mode_val})")
        
        # Categorical specific imputation
        for col in categorical_cols:
            if df_imputed[col].isnull().any() and strategy != 'mode':
                df_imputed[col].fillna('Unknown', inplace=True)
                print(f"   {col}: Filled {self.df[col].isnull().sum()} missing values with 'Unknown'")
        
        self.strategies_applied.append(f"Simple imputation: {strategy}")
        return df_imputed
    
    def advanced_imputation(self, strategy='knn', n_neighbors=5):
        """
        Apply advanced imputation techniques
        """
        from sklearn.impute import KNNImputer, IterativeImputer
        from sklearn.preprocessing import LabelEncoder
        
        df_imputed = self.df.copy()
        
        # Separate numerical and categorical columns
        numerical_cols = df_imputed.select_dtypes(include=[np.number]).columns
        categorical_cols = df_imputed.select_dtypes(include=['object']).columns
        
        # Encode categorical variables for advanced imputation
        label_encoders = {}
        for col in categorical_cols:
            if df_imputed[col].isnull().any():
                le = LabelEncoder()
                # Fit encoder on non-null values
                non_null_values = df_imputed[col].dropna()
                le.fit(non_null_values)
                label_encoders[col] = le
                
                # Transform non-null values
                mask = df_imputed[col].notnull()
                df_imputed.loc[mask, col] = le.transform(df_imputed.loc[mask, col])
        
        # Apply advanced imputation
        if strategy == 'knn':
            imputer = KNNImputer(n_neighbors=n_neighbors)
            imputed_values = imputer.fit_transform(df_imputed.select_dtypes(include=[np.number]))
            
            # Replace values in dataframe
            numerical_cols_with_missing = [col for col in numerical_cols if df_imputed[col].isnull().any()]
            for i, col in enumerate(df_imputed.select_dtypes(include=[np.number]).columns):
                df_imputed[col] = imputed_values[:, i]
            
            print(f"KNN imputation completed with {n_neighbors} neighbors")
            
        elif strategy == 'iterative':
            imputer = IterativeImputer(random_state=42, max_iter=10)
            imputed_values = imputer.fit_transform(df_imputed.select_dtypes(include=[np.number]))
            
            # Replace values in dataframe
            for i, col in enumerate(df_imputed.select_dtypes(include=[np.number]).columns):
                df_imputed[col] = imputed_values[:, i]
            
            print("Iterative imputation completed")
        
        # Decode categorical variables back
        for col, le in label_encoders.items():
            # Round to nearest integer for encoded categorical
            df_imputed[col] = df_imputed[col].round().astype(int)
            # Ensure values are within encoder range
            df_imputed[col] = df_imputed[col].clip(0, len(le.classes_) - 1)
            df_imputed[col] = le.inverse_transform(df_imputed[col])
        
        self.strategies_applied.append(f"Advanced imputation: {strategy}")
        return df_imputed
    
    def forward_fill_backfill(self, method='ffill'):
        """
        Apply forward fill or backward fill (useful for time series)
        """
        df_imputed = self.df.copy()
        
        if method == 'ffill':
            df_imputed = df_imputed.fillna(method='ffill')
            print("Forward fill applied")
        elif method == 'bfill':
            df_imputed = df_imputed.fillna(method='bfill')
            print("Backward fill applied")
        elif method == 'interpolate':
            numerical_cols = df_imputed.select_dtypes(include=[np.number]).columns
            for col in numerical_cols:
                df_imputed[col] = df_imputed[col].interpolate()
            print("Linear interpolation applied to numerical columns")
        
        self.strategies_applied.append(f"Fill method: {method}")
        return df_imputed
    
    def compare_strategies(self, strategies=['mean', 'median', 'knn']):
        """
        Compare different imputation strategies
        """
        results = {}
        
        for strategy in strategies:
            if strategy in ['mean', 'median', 'mode']:
                imputed_df = self.simple_imputation(strategy)
            elif strategy in ['knn', 'iterative']:
                try:
                    imputed_df = self.advanced_imputation(strategy)
                except ImportError:
                    print(f"Scikit-learn required for {strategy} imputation")
                    continue
            
            # Calculate metrics
            numerical_cols = imputed_df.select_dtypes(include=[np.number]).columns
            if len(numerical_cols) > 0:
                results[strategy] = {
                    'mean': imputed_df[numerical_cols].mean().mean(),
                    'std': imputed_df[numerical_cols].std().mean(),
                    'missing_values': imputed_df.isnull().sum().sum()
                }
        
        comparison_df = pd.DataFrame(results).T
        print("\nImputation Strategy Comparison:")
        print(comparison_df.round(3))
        
        return comparison_df
```

### 2.3 Domain-Specific Imputation Strategies

```python
def domain_specific_imputation(df, domain='general'):
    """
    Apply domain-specific imputation strategies
    """
    df_imputed = df.copy()
    
    if domain == 'titanic':
        # Age imputation based on title
        df_imputed['Title'] = df_imputed['sex'].str.extract(r' ([A-Za-z]+)\.')
        
        # Group ages by title for more accurate imputation
        title_age_mapping = df_imputed.groupby('Title')['age'].median()
        
        for title in title_age_mapping.index:
            mask = (df_imputed['Title'] == title) & (df_imputed['age'].isnull())
            df_imputed.loc[mask, 'age'] = title_age_mapping[title]
        
        # Embarked: Use most common port
        most_common_embarked = df_imputed['embarked'].mode().iloc[0]
        df_imputed['embarked'].fillna(most_common_embarked, inplace=True)
        
        # Deck (from cabin): Create 'Unknown' category
        df_imputed['deck'] = df_imputed['deck'].astype(str)
        df_imputed['deck'].fillna('Unknown', inplace=True)
        
        print("Applied Titanic-specific imputation:")
        print("  - Age: Imputed based on passenger title")
        print("  - Embarked: Filled with most common port")
        print("  - Deck: Created 'Unknown' category")
        
        # Drop temporary column
        df_imputed.drop('Title', axis=1, inplace=True)
    
    return df_imputed
```

## üîÑ Part 3: Data Wrangling and Transformation

### 3.1 Data Type Conversions and Standardization

```python
def comprehensive_data_transformation(df):
    """
    Comprehensive data transformation pipeline
    """
    df_transformed = df.copy()
    
    print("DATA TRANSFORMATION PIPELINE")
    print("="*40)
    
    # 1. Date/Time Conversions
    date_columns = ['purchase_date', 'date', 'timestamp']
    for col in df_transformed.columns:
        if any(date_word in col.lower() for date_word in date_columns):
            try:
                df_transformed[col] = pd.to_datetime(df_transformed[col])
                print(f"‚úÖ Converted {col} to datetime")
            except:
                print(f"‚ùå Could not convert {col} to datetime")
    
    # 2. Categorical Conversions
    # Convert low-cardinality object columns to category
    object_cols = df_transformed.select_dtypes(include=['object']).columns
    for col in object_cols:
        unique_ratio = df_transformed[col].nunique() / len(df_transformed)
        if unique_ratio < 0.5:  # Less than 50% unique values
            df_transformed[col] = df_transformed[col].astype('category')
            print(f"‚úÖ Converted {col} to category ({df_transformed[col].nunique()} unique values)")
    
    # 3. Numerical Standardization
    # Convert boolean columns to integers
    bool_cols = df_transformed.select_dtypes(include=['bool']).columns
    for col in bool_cols:
        df_transformed[col] = df_transformed[col].astype(int)
        print(f"‚úÖ Converted boolean {col} to integer")
    
    # 4. Text Standardization
    text_cols = df_transformed.select_dtypes(include=['object']).columns
    for col in text_cols:
        if df_transformed[col].dtype == 'object':
            # Remove extra whitespace
            df_transformed[col] = df_transformed[col].astype(str).str.strip()
            # Standardize case (optional)
            if col.lower() in ['name', 'address', 'description']:
                df_transformed[col] = df_transformed[col].str.title()
            elif col.lower() in ['email', 'url']:
                df_transformed[col] = df_transformed[col].str.lower()
            
            print(f"‚úÖ Standardized text in {col}")
    
    return df_transformed
```

### 4.1 Statistical Methods for Outlier Detection

```python
class OutlierDetector:
    """
    Comprehensive outlier detection and treatment class
    """
    
    def __init__(self, df):
        self.df = df.copy()
        self.numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
        self.outlier_summary = {}
    
    def iqr_method(self, column, multiplier=1.5):
        """
        Detect outliers using Interquartile Range (IQR) method
        """
        Q1 = self.df[column].quantile(0.25)
        Q3 = self.df[column].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - multiplier * IQR
        upper_bound = Q3 + multiplier * IQR
        
        outliers = self.df[(self.df[column] < lower_bound) | (self.df[column] > upper_bound)]
        
        self.outlier_summary[f"{column}_IQR"] = {
            'method': 'IQR',
            'lower_bound': lower_bound,
            'upper_bound': upper_bound,
            'outlier_count': len(outliers),
            'outlier_percentage': len(outliers) / len(self.df) * 100,
            'outlier_indices': outliers.index.tolist()
        }
        
        return outliers
    
    def z_score_method(self, column, threshold=3):
        """
        Detect outliers using Z-score method
        """
        from scipy import stats
        
        z_scores = np.abs(stats.zscore(self.df[column].dropna()))
        outlier_mask = z_scores > threshold
        
        # Get outlier indices from original dataframe
        valid_indices = self.df[column].dropna().index
        outlier_indices = valid_indices[outlier_mask]
        outliers = self.df.loc[outlier_indices]
        
        self.outlier_summary[f"{column}_zscore"] = {
            'method': 'Z-score',
            'threshold': threshold,
            'outlier_count': len(outliers),
            'outlier_percentage': len(outliers) / len(self.df) * 100,
            'outlier_indices': outlier_indices.tolist()
        }
        
        return outliers
    
    def modified_z_score_method(self, column, threshold=3.5):
        """
        Detect outliers using Modified Z-score (based on median)
        """
        median = self.df[column].median()
        mad = np.median(np.abs(self.df[column] - median))
        
        # Avoid division by zero
        if mad == 0:
            mad = np.median(np.abs(self.df[column] - self.df[column].mean()))
        
        modified_z_scores = 0.6745 * (self.df[column] - median) / mad
        outlier_mask = np.abs(modified_z_scores) > threshold
        outliers = self.df[outlier_mask]
        
        self.outlier_summary[f"{column}_modified_zscore"] = {
            'method': 'Modified Z-score',
            'threshold': threshold,
            'outlier_count': len(outliers),
            'outlier_percentage': len(outliers) / len(self.df) * 100,
            'outlier_indices': outliers.index.tolist()
        }
        
        return outliers
    
    def isolation_forest_method(self, columns=None, contamination=0.1):
        """
        Detect outliers using Isolation Forest
        """
        from sklearn.ensemble import IsolationForest
        
        if columns is None:
            columns = self.numerical_cols
        
        # Prepare data
        data_for_isolation = self.df[columns].dropna()
        
        # Fit Isolation Forest
        iso_forest = IsolationForest(contamination=contamination, random_state=42)
        outlier_labels = iso_forest.fit_predict(data_for_isolation)
        
        # Get outlier indices
        outlier_mask = outlier_labels == -1
        outlier_indices = data_for_isolation.index[outlier_mask]
        outliers = self.df.loc[outlier_indices]
        
        self.outlier_summary[f"isolation_forest"] = {
            'method': 'Isolation Forest',
            'contamination': contamination,
            'columns_used': columns,
            'outlier_count': len(outliers),
            'outlier_percentage': len(outliers) / len(self.df) * 100,
            'outlier_indices': outlier_indices.tolist()
        }
        
        return outliers
    
    def detect_all_outliers(self, methods=['iqr', 'zscore', 'modified_zscore']):
        """
        Run multiple outlier detection methods
        """
        print("COMPREHENSIVE OUTLIER DETECTION")
        print("="*50)
        
        for column in self.numerical_cols:
            print(f"\nAnalyzing column: {column}")
            print("-" * 30)
            
            if 'iqr' in methods:
                outliers_iqr = self.iqr_method(column)
                print(f"IQR Method: {len(outliers_iqr)} outliers ({len(outliers_iqr)/len(self.df)*100:.1f}%)")
            
            if 'zscore' in methods:
                try:
                    outliers_zscore = self.z_score_method(column)
                    print(f"Z-score Method: {len(outliers_zscore)} outliers ({len(outliers_zscore)/len(self.df)*100:.1f}%)")
                except:
                    print("Z-score Method: Could not calculate (missing scipy)")
            
            if 'modified_zscore' in methods:
                outliers_modified = self.modified_z_score_method(column)
                print(f"Modified Z-score: {len(outliers_modified)} outliers ({len(outliers_modified)/len(self.df)*100:.1f}%)")
        
        if 'isolation_forest' in methods:
            try:
                outliers_iso = self.isolation_forest_method()
                print(f"\nIsolation Forest (multivariate): {len(outliers_iso)} outliers ({len(outliers_iso)/len(self.df)*100:.1f}%)")
            except ImportError:
                print("\nIsolation Forest: Requires scikit-learn")
    
    def visualize_outliers(self, column, methods=['iqr', 'zscore']):
        """
        Visualize outliers using different methods
        """
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        fig.suptitle(f'Outlier Detection Visualization: {column}', fontsize=16)
        
        # Box plot
        axes[0,0].boxplot(self.df[column].dropna())
        axes[0,0].set_title('Box Plot')
        axes[0,0].set_ylabel('Values')
        
        # Histogram
        axes[0,1].hist(self.df[column].dropna(), bins=30, alpha=0.7, edgecolor='black')
        axes[0,1].set_title('Histogram')
        axes[0,1].set_xlabel('Values')
        axes[0,1].set_ylabel('Frequency')
        
        # Scatter plot with outliers highlighted
        if 'iqr' in methods and f"{column}_IQR" in self.outlier_summary:
            outlier_indices = self.outlier_summary[f"{column}_IQR"]['outlier_indices']
            normal_mask = ~self.df.index.isin(outlier_indices)
            
            axes[1,0].scatter(self.df.index[normal_mask], self.df.loc[normal_mask, column], 
                            alpha=0.6, label='Normal', s=20)
            axes[1,0].scatter(outlier_indices, self.df.loc[outlier_indices, column], 
                            color='red', label='Outliers (IQR)', s=30)
            axes[1,0].set_title('Outliers Highlighted (IQR Method)')
            axes[1,0].set_xlabel('Index')
            axes[1,0].set_ylabel('Values')
            axes[1,0].legend()
        
        # Q-Q plot
        from scipy import stats
        stats.probplot(self.df[column].dropna(), dist="norm", plot=axes[1,1])
        axes[1,1].set_title('Q-Q Plot (Normal Distribution)')
        
        plt.tight_layout()
        plt.show()
    
    def get_outlier_summary(self):
        """
        Get summary of all outlier detection results
        """
        if not self.outlier_summary:
            print("No outlier detection performed yet. Run detect_all_outliers() first.")
            return None
        
        summary_data = []
        for key, info in self.outlier_summary.items():
            summary_data.append({
                'Column_Method': key,
                'Method': info['method'],
                'Outlier_Count': info['outlier_count'],
                'Outlier_Percentage': round(info['outlier_percentage'], 2)
            })
        
        return pd.DataFrame(summary_data)

# Demonstrate outlier detection with sample data
print("OUTLIER DETECTION DEMONSTRATION")
print("="*50)

# Create sample data with outliers
np.random.seed(42)
n_samples = 1000

# Normal data with some outliers
normal_data = np.random.normal(100, 15, int(n_samples * 0.95))
outlier_data = np.random.normal(200, 10, int(n_samples * 0.05))
combined_data = np.concatenate([normal_data, outlier_data])

sample_outlier_df = pd.DataFrame({
    'revenue': combined_data,
    'profit_margin': np.random.normal(0.15, 0.05, n_samples),
    'customer_satisfaction': np.random.beta(8, 2, n_samples) * 100,
    'employee_count': np.random.poisson(50, n_samples)
})

# Add some extreme outliers manually
sample_outlier_df.loc[0, 'revenue'] = 1000  # Extreme high
sample_outlier_df.loc[1, 'profit_margin'] = -0.5  # Extreme low
sample_outlier_df.loc[2, 'customer_satisfaction'] = 10  # Extreme low

print("Sample data with outliers created:")
print(sample_outlier_df.describe())

# Initialize outlier detector
outlier_detector = OutlierDetector(sample_outlier_df)

# Detect outliers using all methods
outlier_detector.detect_all_outliers(['iqr', 'zscore', 'modified_zscore'])

# Get summary
outlier_summary_df = outlier_detector.get_outlier_summary()
print("\nOUTLIER DETECTION SUMMARY:")
print(outlier_summary_df)

# Visualize outliers for revenue column
outlier_detector.visualize_outliers('revenue')
```

### 4.2 Outlier Treatment Strategies

```python
class OutlierTreatment:
    """
    Class for treating outliers using various strategies
    """
    
    def __init__(self, df):
        self.df = df.copy()
        self.treatment_log = []
    
    def remove_outliers(self, column, method='iqr', multiplier=1.5):
        """
        Remove outliers from the dataset
        """
        df_treated = self.df.copy()
        
        if method == 'iqr':
            Q1 = df_treated[column].quantile(0.25)
            Q3 = df_treated[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - multiplier * IQR
            upper_bound = Q3 + multiplier * IQR
            
            # Remove outliers
            outlier_mask = (df_treated[column] < lower_bound) | (df_treated[column] > upper_bound)
            outliers_removed = outlier_mask.sum()
            df_treated = df_treated[~outlier_mask]
            
            self.treatment_log.append(f"Removed {outliers_removed} outliers from {column} using IQR method")
            
        return df_treated
    
    def cap_outliers(self, column, method='iqr', multiplier=1.5, percentiles=(5, 95)):
        """
        Cap outliers at specified boundaries (winsorization)
        """
        df_treated = self.df.copy()
        
        if method == 'iqr':
            Q1 = df_treated[column].quantile(0.25)
            Q3 = df_treated[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - multiplier * IQR
            upper_bound = Q3 + multiplier * IQR
        elif method == 'percentile':
            lower_bound = df_treated[column].quantile(percentiles[0] / 100)
            upper_bound = df_treated[column].quantile(percentiles[1] / 100)
        
        # Count outliers before treatment
        outliers_capped = ((df_treated[column] < lower_bound) | (df_treated[column] > upper_bound)).sum()
        
        # Cap outliers
        df_treated[column] = df_treated[column].clip(lower=lower_bound, upper=upper_bound)
        
        self.treatment_log.append(f"Capped {outliers_capped} outliers in {column} to bounds [{lower_bound:.2f}, {upper_bound:.2f}]")
        
        return df_treated
    
    def transform_outliers(self, column, method='log'):
        """
        Transform data to reduce impact of outliers
        """
        df_treated = self.df.copy()
        
        if method == 'log':
            # Add small constant to handle zeros
            df_treated[f'{column}_log'] = np.log1p(df_treated[column])
            self.treatment_log.append(f"Applied log transformation to {column}")
            
        elif method == 'sqrt':
            df_treated[f'{column}_sqrt'] = np.sqrt(df_treated[column])
            self.treatment_log.append(f"Applied square root transformation to {column}")
            
        elif method == 'box_cox':
            from scipy import stats
            if (df_treated[column] > 0).all():
                df_treated[f'{column}_boxcox'], lambda_param = stats.boxcox(df_treated[column])
                self.treatment_log.append(f"Applied Box-Cox transformation to {column} (lambda={lambda_param:.3f})")
            else:
                print(f"Cannot apply Box-Cox to {column}: contains non-positive values")
        
        return df_treated
    
    def robust_scaling(self, column):
        """
        Apply robust scaling to reduce impact of outliers
        """
        from sklearn.preprocessing import RobustScaler
        
        df_treated = self.df.copy()
        
        scaler = RobustScaler()
        df_treated[f'{column}_robust_scaled'] = scaler.fit_transform(df_treated[[column]])
        
        self.treatment_log.append(f"Applied robust scaling to {column}")
        
        return df_treated, scaler
    
    def compare_treatments(self, column, treatments=['original', 'remove', 'cap', 'log']):
        """
        Compare different outlier treatment methods
        """
        results = {}
        
        for treatment in treatments:
            if treatment == 'original':
                treated_data = self.df[column]
                results[treatment] = {
                    'mean': treated_data.mean(),
                    'median': treated_data.median(),
                    'std': treated_data.std(),
                    'skewness': treated_data.skew(),
                    'count': len(treated_data)
                }
            elif treatment == 'remove':
                df_treated = self.remove_outliers(column)
                treated_data = df_treated[column]
                results[treatment] = {
                    'mean': treated_data.mean(),
                    'median': treated_data.median(),
                    'std': treated_data.std(),
                    'skewness': treated_data.skew(),
                    'count': len(treated_data)
                }
            elif treatment == 'cap':
                df_treated = self.cap_outliers(column)
                treated_data = df_treated[column]
                results[treatment] = {
                    'mean': treated_data.mean(),
                    'median': treated_data.median(),
                    'std': treated_data.std(),
                    'skewness': treated_data.skew(),
                    'count': len(treated_data)
                }
            elif treatment == 'log':
                df_treated = self.transform_outliers(column, 'log')
                treated_data = df_treated[f'{column}_log']
                results[treatment] = {
                    'mean': treated_data.mean(),
                    'median': treated_data.median(),
                    'std': treated_data.std(),
                    'skewness': treated_data.skew(),
                    'count': len(treated_data)
                }
        
        comparison_df = pd.DataFrame(results).T
        return comparison_df.round(3)
    
    def visualize_treatments(self, column, treatments=['original', 'cap', 'log']):
        """
        Visualize the effect of different treatments
        """
        n_treatments = len(treatments)
        fig, axes = plt.subplots(2, n_treatments, figsize=(5*n_treatments, 10))
        
        if n_treatments == 1:
            axes = axes.reshape(2, 1)
        
        for i, treatment in enumerate(treatments):
            if treatment == 'original':
                data = self.df[column]
                title = 'Original Data'
            elif treatment == 'remove':
                df_treated = self.remove_outliers(column)
                data = df_treated[column]
                title = 'Outliers Removed'
            elif treatment == 'cap':
                df_treated = self.cap_outliers(column)
                data = df_treated[column]
                title = 'Outliers Capped'
            elif treatment == 'log':
                df_treated = self.transform_outliers(column, 'log')
                data = df_treated[f'{column}_log']
                title = 'Log Transformed'
            
            # Histogram
            axes[0, i].hist(data, bins=30, alpha=0.7, edgecolor='black')
            axes[0, i].set_title(f'{title}\nHistogram')
            axes[0, i].set_ylabel('Frequency')
            
            # Box plot
            axes[1, i].boxplot(data)
            axes[1, i].set_title(f'{title}\nBox Plot')
            axes[1, i].set_ylabel('Values')
        
        plt.tight_layout()
        plt.show()

# Demonstrate outlier treatment
print("\nOUTLIER TREATMENT DEMONSTRATION")
print("="*50)

# Initialize treatment class
outlier_treatment = OutlierTreatment(sample_outlier_df)

# Compare different treatments for revenue column
print("Comparing outlier treatments for 'revenue' column:")
treatment_comparison = outlier_treatment.compare_treatments('revenue')
print(treatment_comparison)

# Visualize treatments
outlier_treatment.visualize_treatments('revenue', ['original', 'cap', 'log'])

print(f"\nTreatment log:")
for log_entry in outlier_treatment.treatment_log:
    print(f"  - {log_entry}")
```

## üîß Part 5: Feature Engineering and Text Preprocessing
### 5.1 Feature Engineering Techniques

```python
class FeatureEngineer:
    """
    Comprehensive feature engineering class
    """
    
    def __init__(self, df):
        self.df = df.copy()
        self.feature_log = []
    
    def create_datetime_features(self, date_column):
        """
        Extract datetime features from date columns
        """
        df_engineered = self.df.copy()
        
        if df_engineered[date_column].dtype != 'datetime64[ns]':
            df_engineered[date_column] = pd.to_datetime(df_engineered[date_column])
        
        # Extract features
        df_engineered[f'{date_column}_year'] = df_engineered[date_column].dt.year
        df_engineered[f'{date_column}_month'] = df_engineered[date_column].dt.month
        df_engineered[f'{date_column}_day'] = df_engineered[date_column].dt.day
        df_engineered[f'{date_column}_dayofweek'] = df_engineered[date_column].dt.dayofweek
        df_engineered[f'{date_column}_quarter'] = df_engineered[date_column].dt.quarter
        df_engineered[f'{date_column}_is_weekend'] = df_engineered[date_column].dt.dayofweek.isin([5, 6]).astype(int)
        
        # Season feature
        df_engineered[f'{date_column}_season'] = df_engineered[f'{date_column}_month'].map({
            12: 'Winter', 1: 'Winter', 2: 'Winter',
            3: 'Spring', 4: 'Spring', 5: 'Spring',
            6: 'Summer', 7: 'Summer', 8: 'Summer',
            9: 'Fall', 10: 'Fall', 11: 'Fall'
        })
        
        self.feature_log.append(f"Created datetime features from {date_column}")
        return df_engineered
    
    def create_binning_features(self, column, bins=5, method='equal_width'):
        """
        Create binned categorical features from continuous variables
        """
        df_engineered = self.df.copy()
        
        if method == 'equal_width':
            df_engineered[f'{column}_binned'] = pd.cut(df_engineered[column], bins=bins, labels=False)
            df_engineered[f'{column}_binned_labels'] = pd.cut(df_engineered[column], bins=bins)
        elif method == 'equal_frequency':
            df_engineered[f'{column}_binned'] = pd.qcut(df_engineered[column], q=bins, labels=False, duplicates='drop')
            df_engineered[f'{column}_binned_labels'] = pd.qcut(df_engineered[column], q=bins, duplicates='drop')
        elif method == 'custom':
            # Custom bins for age
            if 'age' in column.lower():
                age_bins = [0, 18, 25, 35, 50, 65, 100]
                age_labels = ['Child', 'Young Adult', 'Adult', 'Middle Age', 'Senior', 'Elderly']
                df_engineered[f'{column}_binned_labels'] = pd.cut(df_engineered[column], 
                                                                bins=age_bins, labels=age_labels)
            # Custom bins for income
            elif 'income' in column.lower():
                income_bins = [0, 30000, 50000, 75000, 100000, float('inf')]
                income_labels = ['Low', 'Lower-Middle', 'Middle', 'Upper-Middle', 'High']
                df_engineered[f'{column}_binned_labels'] = pd.cut(df_engineered[column],
                                                                bins=income_bins, labels=income_labels)
        
        self.feature_log.append(f"Created binned features for {column} using {method} method")
        return df_engineered
    
    def create_interaction_features(self, col1, col2, operations=['multiply', 'divide', 'add']):
        """
        Create interaction features between two columns
        """
        df_engineered = self.df.copy()
        
        if 'multiply' in operations:
            df_engineered[f'{col1}_{col2}_multiply'] = df_engineered[col1] * df_engineered[col2]
        
        if 'divide' in operations and (df_engineered[col2] != 0).all():
            df_engineered[f'{col1}_{col2}_divide'] = df_engineered[col1] / df_engineered[col2]
        
        if 'add' in operations:
            df_engineered[f'{col1}_{col2}_add'] = df_engineered[col1] + df_engineered[col2]
        
        if 'subtract' in operations:
            df_engineered[f'{col1}_{col2}_subtract'] = df_engineered[col1] - df_engineered[col2]
        
        self.feature_log.append(f"Created interaction features between {col1} and {col2}")
        return df_engineered
    
    def create_aggregation_features(self, group_col, agg_col, operations=['mean', 'sum', 'count']):
        """
        Create aggregation features grouped by a categorical column
        """
        df_engineered = self.df.copy()
        
        for operation in operations:
            if operation == 'mean':
                agg_feature = df_engineered.groupby(group_col)[agg_col].transform('mean')
                df_engineered[f'{group_col}_{agg_col}_mean'] = agg_feature
            elif operation == 'sum':
                agg_feature = df_engineered.groupby(group_col)[agg_col].transform('sum')
                df_engineered[f'{group_col}_{agg_col}_sum'] = agg_feature
            elif operation == 'count':
                agg_feature = df_engineered.groupby(group_col)[agg_col].transform('count')
                df_engineered[f'{group_col}_{agg_col}_count'] = agg_feature
            elif operation == 'std':
                agg_feature = df_engineered.groupby(group_col)[agg_col].transform('std')
                df_engineered[f'{group_col}_{agg_col}_std'] = agg_feature
        
        self.feature_log.append(f"Created aggregation features: {group_col} grouped by {agg_col}")
        return df_engineered
    
    def create_polynomial_features(self, column, degree=2):
        """
        Create polynomial features
        """
        df_engineered = self.df.copy()
        
        for d in range(2, degree + 1):
            df_engineered[f'{column}_poly_{d}'] = df_engineered[column] ** d
        
        self.feature_log.append(f"Created polynomial features for {column} up to degree {degree}")
        return df_engineered
    
    def create_rolling_features(self, column, windows=[3, 7, 30], operations=['mean', 'std']):
        """
        Create rolling window features (useful for time series)
        """
        df_engineered = self.df.copy()
        
        for window in windows:
            for operation in operations:
                if operation == 'mean':
                    df_engineered[f'{column}_rolling_{window}_mean'] = df_engineered[column].rolling(window=window).mean()
                elif operation == 'std':
                    df_engineered[f'{column}_rolling_{window}_std'] = df_engineered[column].rolling(window=window).std()
                elif operation == 'sum':
                    df_engineered[f'{column}_rolling_{window}_sum'] = df_engineered[column].rolling(window=window).sum()
        
        self.feature_log.append(f"Created rolling window features for {column}")
        return df_engineered

# Demonstrate feature engineering
print("FEATURE ENGINEERING DEMONSTRATION")
print("="*50)

# Create sample dataset for feature engineering
np.random.seed(42)
n_records = 1000

sample_data = pd.DataFrame({
    'customer_id': range(1, n_records + 1),
    'purchase_date': pd.date_range('2023-01-01', periods=n_records, freq='D'),
    'age': np.random.normal(40, 15, n_records),
    'income': np.random.lognormal(10.5, 0.5, n_records),
    'purchase_amount': np.random.gamma(2, 50, n_records),
    'product_category': np.random.choice(['Electronics', 'Clothing', 'Books', 'Home'], n_records),
    'region': np.random.choice(['North', 'South', 'East', 'West'], n_records),
})

# Ensure age is positive
sample_data['age'] = np.abs(sample_data['age'])

print("Original dataset shape:", sample_data.shape)
print("Original columns:", list(sample_data.columns))

# Initialize feature engineer
feature_engineer = FeatureEngineer(sample_data)

# Create datetime features
print("\n1. Creating datetime features...")
sample_with_date_features = feature_engineer.create_datetime_features('purchase_date')
new_date_features = [col for col in sample_with_date_features.columns if 'purchase_date' in col and col != 'purchase_date']
print(f"New datetime features: {new_date_features}")

# Create binning features
print("\n2. Creating binning features...")
sample_with_bins = feature_engineer.create_binning_features('age', method='custom')
sample_with_bins = feature_engineer.create_binning_features('income', method='custom')

# Create interaction features
print("\n3. Creating interaction features...")
sample_with_interactions = feature_engineer.create_interaction_features('age', 'income', ['multiply', 'divide'])

# Create aggregation features
print("\n4. Creating aggregation features...")
sample_with_agg = feature_engineer.create_aggregation_features('product_category', 'purchase_amount', ['mean', 'count'])

# Create polynomial features
print("\n5. Creating polynomial features...")
sample_with_poly = feature_engineer.create_polynomial_features('age', degree=3)

print(f"\nFinal dataset shape: {sample_with_poly.shape}")
print(f"Feature engineering log:")
for log_entry in feature_engineer.feature_log:
    print(f"  - {log_entry}")
```

### 5.2 Text Preprocessing and Analysis

```python
class TextPreprocessor:
    """
    Comprehensive text preprocessing class
    """
    
    def __init__(self):
        self.preprocessing_steps = []
    
    def basic_cleaning(self, text_series):
        """
        Basic text cleaning operations
        """
        import re
        
        cleaned_text = text_series.copy()
        
        # Convert to lowercase
        cleaned_text = cleaned_text.str.lower()
        self.preprocessing_steps.append("Converted to lowercase")
        
        # Remove URLs
        url_pattern = r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
        cleaned_text = cleaned_text.str.replace(url_pattern, '', regex=True)
        self.preprocessing_steps.append("Removed URLs")
        
        # Remove email addresses
        email_pattern = r'\S+@\S+'
        cleaned_text = cleaned_text.str.replace(email_pattern, '', regex=True)
        self.preprocessing_steps.append("Removed email addresses")
        
        # Remove special characters (keep only letters, numbers, spaces)
        cleaned_text = cleaned_text.str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
        self.preprocessing_steps.append("Removed special characters")
        
        # Remove extra whitespace
        cleaned_text = cleaned_text.str.replace(r'\s+', ' ', regex=True).str.strip()
        self.preprocessing_steps.append("Removed extra whitespace")
        
        # Remove very short texts (less than 3 characters)
        cleaned_text = cleaned_text[cleaned_text.str.len() >= 3]
        self.preprocessing_steps.append("Removed texts shorter than 3 characters")
        
        return cleaned_text
    
    def advanced_cleaning(self, text_series, remove_stopwords=True, perform_stemming=False, perform_lemmatization=True):
        """
        Advanced text preprocessing with NLP techniques
        """
        try:
            import nltk
            from nltk.corpus import stopwords
            from nltk.stem import PorterStemmer, WordNetLemmatizer
            from nltk.tokenize import word_tokenize
            
            # Download required NLTK data
            nltk.download('punkt', quiet=True)
            nltk.download('stopwords', quiet=True)
            nltk.download('wordnet', quiet=True)
            nltk.download('averaged_perceptron_tagger', quiet=True)
            
        except ImportError:
            print("NLTK not available. Install with: pip install nltk")
            return text_series
        
        processed_texts = []
        
        # Initialize tools
        stop_words = set(stopwords.words('english')) if remove_stopwords else set()
        stemmer = PorterStemmer() if perform_stemming else None
        lemmatizer = WordNetLemmatizer() if perform_lemmatization else None
        
        for text in text_series:
            if pd.isna(text):
                processed_texts.append('')
                continue
                
            # Tokenization
            tokens = word_tokenize(str(text))
            
            # Remove stopwords
            if remove_stopwords:
                tokens = [word for word in tokens if word not in stop_words]
            
            # Stemming
            if perform_stemming and stemmer:
                tokens = [stemmer.stem(word) for word in tokens]
            
            # Lemmatization
            if perform_lemmatization and lemmatizer:
                tokens = [lemmatizer.lemmatize(word) for word in tokens]
            
            # Join back to text
            processed_text = ' '.join(tokens)
            processed_texts.append(processed_text)
        
        if remove_stopwords:
            self.preprocessing_steps.append("Removed stopwords")
        if perform_stemming:
            self.preprocessing_steps.append("Applied stemming")
        if perform_lemmatization:
            self.preprocessing_steps.append("Applied lemmatization")
        
        return pd.Series(processed_texts, index=text_series.index)
    
    def extract_text_features(self, text_series):
        """
        Extract numerical features from text
        """
        features_df = pd.DataFrame()
        
        # Basic length features
        features_df['text_length'] = text_series.str.len()
        features_df['word_count'] = text_series.str.split().str.len()
        features_df['avg_word_length'] = features_df['text_length'] / features_df['word_count']
        
        # Sentence count (approximation using periods)
        features_df['sentence_count'] = text_series.str.count(r'[.!?]+') + 1
        features_df['avg_sentence_length'] = features_df['word_count'] / features_df['sentence_count']
        
        # Character-based features
        features_df['uppercase_count'] = text_series.str.count(r'[A-Z]')
        features_df['digit_count'] = text_series.str.count(r'\d')
        features_df['punctuation_count'] = text_series.str.count(r'[^\w\s]')
        
        # Ratio features
        features_df['uppercase_ratio'] = features_df['uppercase_count'] / features_df['text_length']
        features_df['digit_ratio'] = features_df['digit_count'] / features_df['text_length']
        features_df['punctuation_ratio'] = features_df['punctuation_count'] / features_df['text_length']
        
        self.preprocessing_steps.append("Extracted text features")
        return features_df
    
    def create_tfidf_features(self, text_series, max_features=100, ngram_range=(1, 2)):
        """
        Create TF-IDF features
        """
        try:
            from sklearn.feature_extraction.text import TfidfVectorizer
        except ImportError:
            print("Scikit-learn required for TF-IDF features")
            return None
        
        # Remove empty strings
        clean_texts = text_series.fillna('').astype(str)
        
        # Create TF-IDF vectorizer
        tfidf = TfidfVectorizer(
            max_features=max_features,
            ngram_range=ngram_range,
            stop_words='english'
        )
        
        # Fit and transform
        tfidf_matrix = tfidf.fit_transform(clean_texts)
        
        # Convert to DataFrame
        feature_names = tfidf.get_feature_names_out()
        tfidf_df = pd.DataFrame(
            tfidf_matrix.toarray(), 
            columns=[f'tfidf_{name}' for name in feature_names],
            index=text_series.index
        )
        
        self.preprocessing_steps.append(f"Created {max_features} TF-IDF features")
        return tfidf_df, tfidf
    
    def sentiment_analysis(self, text_series):
        """
        Perform basic sentiment analysis
        """
        try:
            from textblob import TextBlob
        except ImportError:
            print("TextBlob required for sentiment analysis. Install with: pip install textblob")
            return None
        
        sentiments = []
        for text in text_series:
            if pd.isna(text) or text == '':
                sentiments.append({'polarity': 0, 'subjectivity': 0})
            else:
                blob = TextBlob(str(text))
                sentiments.append({
                    'polarity': blob.sentiment.polarity,
                    'subjectivity': blob.sentiment.subjectivity
                })
        
        sentiment_df = pd.DataFrame(sentiments, index=text_series.index)
        
        # Add categorical sentiment
        sentiment_df['sentiment_category'] = sentiment_df['polarity'].apply(
            lambda x: 'positive' if x > 0.1 else 'negative' if x < -0.1 else 'neutral'
        )
        
        self.preprocessing_steps.append("Performed sentiment analysis")
        return sentiment_df

# Demonstrate text preprocessing
print("\nTEXT PREPROCESSING DEMONSTRATION")
print("="*50)

# Create sample text data (simulating customer reviews)
sample_reviews = [
    "This product is absolutely AMAZING!!! I love it so much. Best purchase ever! üòç",
    "terrible quality... waste of money. would not recommend to anyone.",
    "It's okay, nothing special. Average product for the price.",
    "EXCELLENT customer service! Fast shipping and great quality. 5 stars!",
    "Product arrived damaged. Very disappointed with this purchase.",
    "Good value for money. Works as expected. Would buy again.",
    "Outstanding! Exceeded my expectations in every way possible!",
    "Poor design and cheap materials. Save your money and buy something else.",
    "Decent product but shipping was slow. Otherwise satisfied.",
    "Perfect! Exactly what I was looking for. Highly recommended!",
    "Not bad, but could be better. Average quality and service.",
    "Horrible experience. Product broke after one day. Terrible!",
    "Great product! Fast delivery and excellent packaging. Very happy!",
    "Mediocre at best. Nothing impressive about this product.",
    "Fantastic quality and great price! Will definitely order again!"
]

# Create DataFrame with text data
text_df = pd.DataFrame({
    'review_id': range(1, len(sample_reviews) + 1),
    'review_text': sample_reviews,
    'product_category': np.random.choice(['Electronics', 'Clothing', 'Books'], len(sample_reviews)),
    'rating': np.random.choice([1, 2, 3, 4, 5], len(sample_reviews))
})

print("Sample text data:")
print(text_df.head())

# Initialize text preprocessor
text_processor = TextPreprocessor()

# Basic cleaning
print("\n1. Basic text cleaning...")
cleaned_reviews = text_processor.basic_cleaning(text_df['review_text'])
print("Sample cleaned text:")
print(cleaned_reviews.head(3).values)

# Advanced cleaning (if NLTK is available)
print("\n2. Advanced text preprocessing...")
try:
    advanced_cleaned = text_processor.advanced_cleaning(cleaned_reviews, 
                                                       remove_stopwords=True, 
                                                       perform_lemmatization=True)
    print("Sample advanced cleaned text:")
    print(advanced_cleaned.head(3).values)
except:
    print("NLTK not available - skipping advanced preprocessing")
    advanced_cleaned = cleaned_reviews

# Extract text features
print("\n3. Extracting text features...")
text_features = text_processor.extract_text_features(text_df['review_text'])
print("Text features:")
print(text_features.head())

# TF-IDF features (if scikit-learn is available)
print("\n4. Creating TF-IDF features...")
try:
    tfidf_features, tfidf_vectorizer = text_processor.create_tfidf_features(advanced_cleaned, max_features=20)
    print(f"TF-IDF features shape: {tfidf_features.shape}")
    print("Top TF-IDF features:")
    print(tfidf_features.columns[:10].tolist())
except:
    print("Scikit-learn not available - skipping TF-IDF features")

# Sentiment analysis (if TextBlob is available)
print("\n5. Sentiment analysis...")
try:
    sentiment_features = text_processor.sentiment_analysis(text_df['review_text'])
    print("Sentiment analysis results:")
    print(sentiment_features.head())
    
    # Show sentiment distribution
    print("\nSentiment distribution:")
    print(sentiment_features['sentiment_category'].value_counts())
except:
    print("TextBlob not available - skipping sentiment analysis")

print(f"\nText preprocessing steps applied:")
for step in text_processor.preprocessing_steps:
    print(f"  - {step}")
```

```python
def exercise_2_multi_dataset_integration():
    """
    Exercise 2: Integrate multiple datasets with different challenges
    """
    print("\nEXERCISE 2: MULTI-DATASET INTEGRATION")
    print("="*60)
    
    # Create realistic business scenario datasets
    
    # Dataset 1: Customer Information
    np.random.seed(42)
    customers = pd.DataFrame({
        'customer_id': range(1, 1001),
        'first_name': [f"FirstName{i}" for i in range(1, 1001)],
        'last_name': [f"LastName{i}" for i in range(1, 1001)],
        'email': [f"customer{i}@email.com" for i in range(1, 1001)],
        'phone': [f"555-{i:04d}" for i in range(1, 1001)],
        'registration_date': pd.date_range('2020-01-01', periods=1000, freq='D'),
        'age': np.random.normal(40, 15, 1000),
        'gender': np.random.choice(['M', 'F', 'Other'], 1000, p=[0.45, 0.45, 0.1]),
        'income_range': np.random.choice(['<30K', '30-50K', '50-75K', '75-100K', '>100K'], 1000),
        'city': np.random.choice(['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'], 1000)
    })
    
    # Add missing values
    customers.loc[customers.sample(50).index, 'age'] = np.nan
    customers.loc[customers.sample(30).index, 'income_range'] = np.nan
    
    # Dataset 2: Transactions (some customers missing)
    transaction_customers = np.random.choice(range(1, 1201), 5000)  # Some customers not in main table
    transactions = pd.DataFrame({
        'transaction_id': range(1, 5001),
        'customer_id': transaction_customers,
        'transaction_date': pd.date_range('2022-01-01', periods=5000, freq='H'),
        'product_category': np.random.choice(['Electronics', 'Clothing', 'Books', 'Home', 'Sports'], 5000),
        'amount': np.random.gamma(2, 50, 5000),
        'payment_method': np.random.choice(['Credit Card', 'Debit Card', 'PayPal', 'Cash'], 5000),
        'store_location': np.random.choice(['Online', 'Store A', 'Store B', 'Store C'], 5000)
    })
    
    # Add missing values and data quality issues
    transactions.loc[transactions.sample(100).index, 'amount'] = np.nan
    transactions.loc[transactions.sample(50).index, 'product_category'] = ''
    
    # Dataset 3: Product Information
    products = pd.DataFrame({
        'product_category': ['Electronics', 'Clothing', 'Books', 'Home', 'Sports'],
        'avg_price': [299.99, 79.99, 19.99, 149.99, 89.99],
        'profit_margin': [0.15, 0.35, 0.25, 0.20, 0.30],
        'return_rate': [0.08, 0.12, 0.03, 0.06, 0.09]
    })
    
    print("Datasets created:")
    print(f"Customers: {customers.shape}")
    print(f"Transactions: {transactions.shape}")
    print(f"Products: {products.shape}")
    
    # Step 1: Individual dataset cleaning
    print("\n--- Step 1: Individual Dataset Cleaning ---")
    
    # Clean customers
    print("Cleaning customers dataset...")
    customers_clean = customers.copy()
    
    # Handle missing age with median by gender
    for gender in customers_clean['gender'].unique():
        mask = (customers_clean['gender'] == gender) & (customers_clean['age'].isnull())
        median_age = customers_clean[customers_clean['gender'] == gender]['age'].median()
        customers_clean.loc[mask, 'age'] = median_age
    
    # Handle missing income_range with mode
    mode_income = customers_clean['income_range'].mode().iloc[0]
    customers_clean['income_range'].fillna(mode_income, inplace=True)
    
    # Clean transactions
    print("Cleaning transactions dataset...")
    transactions_clean = transactions.copy()
    
    # Handle missing amounts with median by category
    for category in transactions_clean['product_category'].unique():
        if category != '':
            mask = (transactions_clean['product_category'] == category) & (transactions_clean['amount'].isnull())
            median_amount = transactions_clean[transactions_clean['product_category'] == category]['amount'].median()
            transactions_clean.loc[mask, 'amount'] = median_amount
    
    # Handle empty product categories
    transactions_clean['product_category'].replace('', 'Unknown', inplace=True)
    
    # Handle remaining missing amounts
    transactions_clean['amount'].fillna(transactions_clean['amount'].median(), inplace=True)
    
    # Step 2: Data Integration
    print("\n--- Step 2: Data Integration ---")
    
    # Join customers and transactions
    customer_transactions = pd.merge(transactions_clean, customers_clean, on='customer_id', how='left')
    
    # Add product information
    customer_transactions = pd.merge(customer_transactions, products, on='product_category', how='left')
    
    print(f"Integrated dataset shape: {customer_transactions.shape}")
    
    # Identify customers without demographic data
    missing_demographics = customer_transactions['age'].isnull().sum()
    print(f"Transactions without customer demographics: {missing_demographics}")
    
    # Step 3: Feature Engineering on Integrated Data
    print("\n--- Step 3: Feature Engineering ---")
    
    feature_engineer = FeatureEngineer(customer_transactions)
    
    # Create datetime features
    integrated_with_features = feature_engineer.create_datetime_features('transaction_date')
    integrated_with_features = feature_engineer.create_datetime_features('registration_date')
    
    # Create customer tenure feature
    integrated_with_features['customer_tenure_days'] = (
        integrated_with_features['transaction_date'] - integrated_with_features['registration_date']
    ).dt.days
    
    # Create aggregated customer features
    customer_summary = integrated_with_features.groupby('customer_id').agg({
        'amount': ['sum', 'mean', 'count', 'std'],
        'transaction_date': ['min', 'max'],
        'product_category': lambda x: x.nunique()
    }).round(2)
    
    # Flatten column names
    customer_summary.columns = ['_'.join(col).strip() for col in customer_summary.columns]
    customer_summary = customer_summary.reset_index()
    
    # Rename columns for clarity
    customer_summary.rename(columns={
        'amount_sum': 'total_spent',
        'amount_mean': 'avg_transaction_amount',
        'amount_count': 'transaction_count',
        'amount_std': 'spending_variability',
        'transaction_date_min': 'first_transaction',
        'transaction_date_max': 'last_transaction',
        'product_category_<lambda>': 'categories_purchased'
    }, inplace=True)
    
    print(f"Customer summary features created: {customer_summary.shape}")
    print("Sample customer summary:")
    print(customer_summary.head())
    
    # Step 4: Final Integration and Quality Check
    print("\n--- Step 4: Final Integration and Quality Check ---")
    
    # Merge back customer summary features
    final_dataset = pd.merge(integrated_with_features, customer_summary, on='customer_id', how='left')
    
    print(f"Final integrated dataset shape: {final_dataset.shape}")
    print(f"Final missing values: {final_dataset.isnull().sum().sum()}")
    
    # Create data quality report for final dataset
    final_missing, final_dtypes = comprehensive_data_quality_report(final_dataset, "Final Integrated Dataset")
    
    return final_dataset, customer_summary

# Run Exercise 2
exercise_2_result, customer_features = exercise_2_multi_dataset_integration()
```

```python
def exercise_3_text_data_challenge():
    """
    Exercise 3: Complex text data preparation scenario
    """
    print("\nEXERCISE 3: TEXT DATA PREPARATION CHALLENGE")
    print("="*60)
    
    # Create realistic customer feedback dataset
    sample_feedback = [
        "ABSOLUTELY TERRIBLE!!! Product arrived broken and customer service was useless. Waste of $150!",
        "Great product! Fast shipping and excellent quality. Would definitely recommend to friends. 5/5 stars!",
        "Product is okay... nothing special. Delivery was slow (took 2 weeks) but item works as expected.",
        "LOVE IT!!! üòçüòçüòç Best purchase ever! Amazing quality and super fast delivery. Thank you so much!",
        "Disappointed with this purchase. Product description was misleading. Returning immediately.",
        "Good value for the money. Not the best quality but works fine for the price point. Average rating.",
        "Outstanding customer service! Had an issue and they resolved it within 24 hours. Very impressed!",
        "Poor quality materials. Product feels cheap and flimsy. Would not buy again or recommend.",
        "Perfect size and color! Exactly what I was looking for. Fast shipping and great packaging.",
        "Product broke after 3 days of use. Very disappointed. Expensive for such poor quality.",
        "Excellent! Exceeded my expectations in every way. Fast delivery, great quality, fair price.",
        "Meh... it's fine I guess. Nothing exciting but does what it's supposed to do. Average product.",
        "HORRIBLE EXPERIENCE! Wrong item sent twice! Customer service is incompetent! Avoid this seller!",
        "Really happy with this purchase! Great quality and arrived earlier than expected. Highly recommend!",
        "Product is decent but overpriced. You can find similar items for half the price elsewhere.",
        "Amazing quality! Beautiful design and excellent craftsmanship. Worth every penny spent!",
        "Terrible packaging! Item arrived damaged due to poor shipping materials. Very frustrating!",
        "Good product overall. Minor issues with setup but customer support helped resolve quickly.",
        "Not as described. Color was different and size was smaller than advertised. Returning.",
        "BEST SELLER EVER! Fast responses, quick shipping, and excellent product quality. A+++"
    ]
    
    # Create comprehensive feedback dataset
    np.random.seed(42)
    feedback_df = pd.DataFrame({
        'feedback_id': range(1, len(sample_feedback) + 1),
        'customer_id': np.random.randint(1000, 9999, len(sample_feedback)),
        'product_id': np.random.choice(['PROD_A', 'PROD_B', 'PROD_C', 'PROD_D', 'PROD_E'], len(sample_feedback)),
        'feedback_text': sample_feedback,
        'rating': np.random.choice([1, 2, 3, 4, 5], len(sample_feedback), p=[0.1, 0.1, 0.2, 0.3, 0.3]),
        'feedback_date': pd.date_range('2023-01-01', periods=len(sample_feedback), freq='D'),
        'feedback_source': np.random.choice(['Website', 'Email', 'Phone', 'Social Media'], len(sample_feedback)),
        'customer_type': np.random.choice(['New', 'Returning', 'VIP'], len(sample_feedback))
    })
    
    print("Sample feedback dataset:")
    print(feedback_df.head())
    print(f"Dataset shape: {feedback_df.shape}")
    
    # Step 1: Text Preprocessing Pipeline
    print("\n--- Step 1: Text Preprocessing Pipeline ---")
    
    text_processor = TextPreprocessor()
    
    # Basic cleaning
    cleaned_feedback = text_processor.basic_cleaning(feedback_df['feedback_text'])
    print("Sample cleaned feedback:")
    for i in range(3):
        print(f"Original: {feedback_df['feedback_text'].iloc[i]}")
        print(f"Cleaned:  {cleaned_feedback.iloc[i]}")
        print()
    
    # Advanced preprocessing (if available)
    try:
        advanced_cleaned = text_processor.advanced_cleaning(cleaned_feedback)
        feedback_df['processed_text'] = advanced_cleaned
    except:
        feedback_df['processed_text'] = cleaned_feedback
        print("Advanced preprocessing not available - using basic cleaning")
    
    # Step 2: Text Feature Extraction
    print("\n--- Step 2: Text Feature Extraction ---")
    
    # Extract basic text features
    text_features = text_processor.extract_text_features(feedback_df['feedback_text'])
    
    # Combine with original dataset
    feedback_with_features = pd.concat([feedback_df, text_features], axis=1)
    
    print("Text features extracted:")
    print(text_features.columns.tolist())
    
    # Step 3: Sentiment Analysis
    print("\n--- Step 3: Sentiment Analysis ---")
    
    try:
        sentiment_features = text_processor.sentiment_analysis(feedback_df['feedback_text'])
        feedback_with_features = pd.concat([feedback_with_features, sentiment_features], axis=1)
        
        print("Sentiment distribution:")
        print(sentiment_features['sentiment_category'].value_counts())
        
        # Compare sentiment with actual ratings
        sentiment_rating_comparison = pd.crosstab(
            feedback_with_features['sentiment_category'], 
            feedback_with_features['rating']
        )
        print("\nSentiment vs Rating Cross-tabulation:")
        print(sentiment_rating_comparison)
        
    except:
        print("Sentiment analysis not available")
    
    # Step 4: Advanced Text Analysis
    print("\n--- Step 4: Advanced Text Analysis ---")
    
    # Create TF-IDF features
    try:
        tfidf_features, tfidf_vectorizer = text_processor.create_tfidf_features(
            feedback_df['processed_text'], 
            max_features=50
        )
        
        print(f"TF-IDF features shape: {tfidf_features.shape}")
        
        # Find most important terms by sentiment
        if 'sentiment_category' in feedback_with_features.columns:
            for sentiment in ['positive', 'negative', 'neutral']:
                sentiment_mask = feedback_with_features['sentiment_category'] == sentiment
                if sentiment_mask.sum() > 0:
                    avg_tfidf = tfidf_features[sentiment_mask].mean()
                    top_terms = avg_tfidf.nlargest(5)
                    print(f"\nTop terms for {sentiment} sentiment:")
                    for term, score in top_terms.items():
                        clean_term = term.replace('tfidf_', '')
                        print(f"  {clean_term}: {score:.3f}")
    
    except:
        print("TF-IDF analysis not available")
    
    # Step 5: Feature Engineering for Text Data
    print("\n--- Step 5: Feature Engineering for Text Data ---")
    
    # Create custom text-based features
    feedback_with_features['has_exclamation'] = feedback_df['feedback_text'].str.contains('!').astype(int)
    feedback_with_features['has_question'] = feedback_df['feedback_text'].str.contains(r'\?').astype(int)
    feedback_with_features['has_caps'] = feedback_df['feedback_text'].str.contains(r'[A-Z]{3,}').astype(int)
    feedback_with_features['has_emoji'] = feedback_df['feedback_text'].str.contains(r'[üòÄ-üôè]').astype(int)
    feedback_with_features['has_price'] = feedback_df['feedback_text'].str.contains(r'\$\d+').astype(int)
    
    # Extract mentioned numbers (could be prices, ratings, etc.)
    feedback_with_features['mentioned_numbers'] = feedback_df['feedback_text'].str.findall(r'\d+').str.len()
    
    # Create urgency indicators
    urgency_words = ['urgent', 'immediately', 'asap', 'quickly', 'fast', 'slow', 'late', 'delayed']
    feedback_with_features['urgency_score'] = feedback_df['feedback_text'].str.lower().str.count('|'.join(urgency_words))
    
    # Create quality indicators
    quality_positive = ['excellent', 'outstanding', 'amazing', 'perfect', 'great', 'good', 'quality']
    quality_negative = ['terrible', 'horrible', 'awful', 'poor', 'bad', 'cheap', 'broken']
    
    feedback_with_features['positive_quality_words'] = feedback_df['feedback_text'].str.lower().str.count('|'.join(quality_positive))
    feedback_with_features['negative_quality_words'] = feedback_df['feedback_text'].str.lower().str.count('|'.join(quality_negative))
    feedback_with_features['quality_score'] = feedback_with_features['positive_quality_words'] - feedback_with_features['negative_quality_words']
    
    print("Custom text features created:")
    custom_features = ['has_exclamation', 'has_question', 'has_caps', 'has_emoji', 'has_price', 
                      'mentioned_numbers', 'urgency_score', 'quality_score']
    print(custom_features)
    
    # Step 6: Final Analysis and Insights
    print("\n--- Step 6: Final Analysis and Insights ---")
    
    # Correlation between text features and ratings
    if len(feedback_with_features.select_dtypes(include=[np.number]).columns) > 0:
        correlation_with_rating = feedback_with_features.select_dtypes(include=[np.number]).corr()['rating'].abs().sort_values(ascending=False)
        print("Features most correlated with rating:")
        print(correlation_with_rating.head(10))
    
    # Summary statistics by product
    product_summary = feedback_with_features.groupby('product_id').agg({
        'rating': ['mean', 'count'],
        'word_count': 'mean',
        'quality_score': 'mean',
        'urgency_score': 'mean'
    }).round(2)
    
    product_summary.columns = ['_'.join(col).strip() for col in product_summary.columns]
    print(f"\nProduct summary:")
    print(product_summary)
    
    print(f"\nFinal dataset shape: {feedback_with_features.shape}")
    print(f"Text preprocessing steps applied:")
    for step in text_processor.preprocessing_steps:
        print(f"  - {step}")
    
    return feedback_with_features
```

### Data Quality Validation Framework

```python
def validate_data_preparation(original_df, processed_df, documentation_dict):
    """
    Validate the quality of data preparation work
    """
    print("DATA PREPARATION VALIDATION")
    print("="*40)
    
    validation_results = {}
    
    # 1. Completeness Check
    original_missing = original_df.isnull().sum().sum()
    processed_missing = processed_df.isnull().sum().sum()
    missing_reduction = ((original_missing - processed_missing) / original_missing) * 100 if original_missing > 0 else 100
    
    validation_results['missing_data_handled'] = {
        'original_missing': original_missing,
        'processed_missing': processed_missing,
        'reduction_percentage': missing_reduction,
        'passed': processed_missing < original_missing or original_missing == 0
    }
    
    # 2. Data Integrity Check
    # Check if data types are appropriate
    numerical_cols = processed_df.select_dtypes(include=[np.number]).columns
    categorical_cols = processed_df.select_dtypes(include=['object', 'category']).columns
    
    validation_results['data_types'] = {
        'numerical_columns': len(numerical_cols),
        'categorical_columns': len(categorical_cols),
        'appropriate_types': True  # This would need domain-specific validation
    }
    
    # 3. Documentation Quality Check
    required_documentation = [
        'missing_data_strategy',
        'outlier_treatment',
        'feature_engineering',
        'data_transformations'
    ]
    
    doc_completeness = sum([key in documentation_dict for key in required_documentation]) / len(required_documentation)
    
    validation_results['documentation'] = {
        'completeness_score': doc_completeness,
        'passed': doc_completeness >= 0.8
    }
    
    # 4. Feature Engineering Validation
    original_features = len(original_df.columns)
    processed_features = len(processed_df.columns)
    features_added = processed_features - original_features
    
    validation_results['feature_engineering'] = {
        'original_features': original_features,
        'processed_features': processed_features,
        'features_added': features_added,
        'engineering_performed': features_added > 0
    }
    
    # Print validation results
    print("Validation Results:")
    print("-" * 20)
    
    for category, results in validation_results.items():
        print(f"\n{category.upper()}:")
        for metric, value in results.items():
            if isinstance(value, bool):
                status = "‚úÖ PASS" if value else "‚ùå FAIL"
                print(f"  {metric}: {status}")
            else:
                print(f"  {metric}: {value}")
    
    # Overall score
    pass_count = sum([
        validation_results['missing_data_handled']['passed'],
        validation_results['documentation']['passed'],
        validation_results['feature_engineering']['engineering_performed']
    ])
    
    overall_score = (pass_count / 3) * 100
    print(f"\nOVERALL VALIDATION SCORE: {overall_score:.1f}%")
    
    if overall_score >= 80:
        print("üéâ EXCELLENT data preparation work!")
    elif overall_score >= 60:
        print("üëç GOOD data preparation work with room for improvement")
    else:
        print("‚ö†Ô∏è  Data preparation needs significant improvement")
    
    return validation_results

# Example validation (you would run this on your actual work)
print("\nEXAMPLE VALIDATION:")
sample_documentation = {
    'missing_data_strategy': 'Used domain-specific imputation for Titanic dataset',
    'outlier_treatment': 'Applied IQR method with capping for outliers',
    'feature_engineering': 'Created age bins, family size, and title features',
    'data_transformations': 'Applied log transformation to fare column'
}
```
